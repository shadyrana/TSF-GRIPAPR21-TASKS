This task is based on supervised learing in which I have used Linear Regression.
I have also used libraries like Numpy, Pandas, Seaborn, Matplotlib, and sklearn. 
Objective: To predict the marks obtained by a student if he studies 9.25 hrs.

# Importing all the required libraries:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split 
```

# Reading data from link:

```python
url = "http://bit.ly/w-data"
df = pd.read_csv(url)
print("Data imported successfully!")
```

# Visualizing the data:

```python
df.plot(x='Hours', y='Scores', style='o', figsize=(12, 7))      #Using the plot() to visualize data
plt.title('Hours vs Percentage',color="green", size=25)      
plt.xlabel('Hours Studied', color="purple",size=20)                    
plt.ylabel('Percentage Score', color="purple",size=20)                
```

# Data Preprocessing:

```python
x = df.iloc[:, :-1].values    
y = df.iloc[:, 1].values
```

# Model Training:

```python
from sklearn.linear_model import LinearRegression
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0) 
regressor = LinearRegression()  
regressor.fit(x_train, y_train) 
print("The training of model is complete!")
```

# Plotting line of regression for test data:

```python
line = regressor.coef_*x+regressor.intercept_

plt.figure(figsize=(12, 7))
plt.scatter(x, y)
plt.plot(x, line, color="purple")
plt.show()
```

```python
print(x_test)              #Testing the data in hours
y_pred = regressor.predict(x_test)
```

```python
dfAP = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
dfAP                #Comparing Actual vs Predicted values
```

```python
print("Training Score: ",regressor.score(x_train, y_train))
print("Testing Score: ",regressor.score(x_test, y_test))
```

# Predicting Values:

```python
s = float(input("Enter the Number of Hours: "))
self_prediction = regressor.predict(np.array([s]).reshape(-1,1)) #Predicting the value using regressor    
print("Number of Hours = {} \nPredicted Score = {}".format(s,self_prediction[0]))
```

# Final Evaluation:

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, accuracy_score
print("Mean Squared Error: ",mean_squared_error(y_test, y_pred))
print("Mean Absolute Error: ",mean_absolute_error(y_test, y_pred)) 
print("Root Mean Squared Error: ",np.sqrt(mean_absolute_error(y_test, y_pred)))
print("R-2 Score: ",r2_score(y_test, y_pred))
```
